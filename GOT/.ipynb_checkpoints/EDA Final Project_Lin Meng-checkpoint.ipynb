```python
import pandas as pd
import numpy as np
```

```python
## load the scripts data
scripts = pd.read_csv("got_scripts.csv", sep = ";", parse_dates = ["Date"], header = 0
                     , names = ["Row_num", "Season", "Episode_name", "Sentence", "First_name", "N_series", "Episode", "Date"])
scripts.head(10)
```

```python
## drop the NAs
main_scripts = scripts.dropna()
main_scripts.head()
```

## 2. Most Common Words

```python
import nltk
from nltk.stem.porter import *
from sklearn.feature_extraction import stop_words
from collections import Counter
from collections import defaultdict
import string
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def tokenize(text):
    """
    Tokenize text and return a non-unique list of tokenized words
    found in the text. Normalize to lowercase, strip punctuation,
    remove stop words, drop words of length < 3, strip digits.
    """
    text = text.lower()
    text = re.sub("[" + string.punctuation + "0-9\\r\\t\\n]", " ", text)
    tokens = nltk.word_tokenize(text)
    tokens = [w for w in tokens if len(w) > 2]
    return [w for w in tokens if w not in stop_words.ENGLISH_STOP_WORDS]

def stemwords(words):
    """
    Given a list of tokens/words, return a new list with each word
    stemmed using a PorterStemmer.
    """
    stemmer = PorterStemmer()
    return [stemmer.stem(w) for w in words]
```

```python
## per character
words_dict = defaultdict(list)
for char in main_scripts.Character.unique():
    words = main_scripts.loc[main_scripts.Character == char]["Sentence"]
    words = " ".join([sentence for sentence in words])
    words_dict[char] = Counter(stemwords(tokenize(words))).most_common(20)
```

```python
## per gender
words_dict = defaultdict(list)
for gend in main_scripts.Gender.unique():
    words = main_scripts.loc[main_scripts.Gender == gend]["Sentence"]
    words = " ".join([sentence for sentence in words])
    words_dict[gend] = Counter(stemwords(tokenize(words))).most_common(20)
```

```python
## result
who = input()
most_common = pd.DataFrame(words_dict[who], columns = ["Word", "Count"])
print(most_common)
```

## 3. Sentiment Analysis

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
## compute sentiment score
analyzer = SentimentIntensityAnalyzer()
sentences = main_scripts.Sentence
vader_score = []
for sentence in sentences:
    vs = analyzer.polarity_scores(sentence)["compound"]
    vader_score.append(vs)
# print(vader_score)
```

```python
## add to the dataset
main_scripts = main_scripts.reset_index()
main_scripts.loc[:, "Vader_score"] = pd.Series(vader_score)
```

```python
## aggregate the vader scores
## character
vscore_char = main_scripts.groupby(["Season", "Episode", "N_series", "Character", "First_name"]).agg({"Vader_score":"mean"}).reset_index()
vscore_char["SE"] = vscore_char.Season + "-" + vscore_char.Episode.astype(str)
## gender
vscore_gend = main_scripts.groupby(["Season", "Episode", "N_series", "Gender"]).agg({"Vader_score":"mean"}).reset_index()
vscore_gend["SE"] = vscore_gend.Season + "-" + vscore_gend.Episode.astype(str)
```

```python
## compute moving average score
win = 5
male = vscore_gend.loc[vscore_gend.Gender == "M"]
male.loc[:, "MA_score"] = male["Vader_score"].rolling(win, center = True).mean()
female = vscore_gend.loc[vscore_gend.Gender == "F"]
female.loc[:, "MA_score"] = female["Vader_score"].rolling(win, center = True).mean()
# vscore_gend = pd.concat([male, female], axis = 0)
# vscore_gend
```

```python
## factor the Season-Episode
male = male.sort_values(by = "N_series", ascending = True)
male.SE = male.SE.astype(pd.api.types.CategoricalDtype(categories=list(male.SE), ordered = True), )
female = female.sort_values(by = "N_series", ascending = True)
female.SE = female.SE.astype(pd.api.types.CategoricalDtype(categories=list(female.SE), ordered = True), )
```

```python
# vscore_tot = main_scripts.groupby(["Character", "First_name", "Gender"]).agg({"Vader_score":"mean"}).reset_index()
# # vscore_tot = vscore_tot.sort_values(by = ["Gender", "Vader_score"], ascending = [True, False])
# vscore_tot = vscore_tot.sort_values(by = "Vader_score", ascending = False)
# vscore_tot.Character = vscore_tot.Character.astype(pd.api.types.CategoricalDtype(categories=list(vscore_tot.Character), ordered = True), )
```

```python
## summarize core characters
vscore_core = main_scripts.loc[main_scripts.First_name.isin(core_char)]
vscore_core = vscore_core.groupby(["Character", "First_name", "Gender"]).agg({"Vader_score":"mean"}).reset_index()
vscore_core = vscore_core.sort_values(by = "Vader_score", ascending = True)
vscore_core.Character = vscore_core.Character.astype(pd.api.types.CategoricalDtype(categories=list(vscore_core.Character), ordered = True), )
vscore_core.Vader_score = vscore_core.Vader_score.apply(round, args = (2,))
```
